```
df_datas_1 = spark.createDataFrame(["2021-07-05T10:00:00.000+0000","2022-12-05T00:09:00.000+0000","2017-02-23T16:23:00.000+0000"],"string").toDF("datas")
df_datas_2 = spark.createDataFrame(["2021-07-05 10:00","2022-12-05 00:09","2017-02-23 16:23"],"string").toDF("datas")
df_datas_3 = spark.createDataFrame(["05/07/2021","05/12/2020","23/02/2017"],"string").toDF("datas")
df_datas_4 = spark.createDataFrame(["05/07/2021 10:30","05/12/2020 14:15","23/02/2017 20:00"],"string").toDF("datas")

display(df_datas_1)
display(df_datas_2)
display(df_datas_3)
display(df_datas_4)
```

```
df_datas_1.createOrReplaceTempView("datas_1")
df_datas_2.createOrReplaceTempView("datas_2")
df_datas_3.createOrReplaceTempView("datas_3")
df_datas_4.createOrReplaceTempView("datas_4")
```

```
%sql
/*Transformar somente DATA -> CAST(xxx as date)*/
select
  cast(datas as date)
from datas_1
```

```
%sql
/*Transformar DATA e HORA -> TO_TIMESTAMP(#DF) 
 # -> onde o DF Ã© o xxx - contido no to_DF("xxx") declarado*/

select
  TO_TIMESTAMP(datas) DATAS
from datas_1
```

```
df_datas_sql = spark.sql("""
 select TO_TIMESTAMP(datas)
 from datas_1
""")

display(df_datas_sql)

df_datas_sql2 = spark.sql("""
    select
    cast(datas as date)
    from datas_2
""")

display(df_datas_sql2)

df_datas_sql3 = spark.sql("""
    select
    to_date(datas, "dd/MM/yyyy")
    from datas_3
""")

display(df_datas_sql3)

df_datas_sql4 = spark.sql("""
    select
    to_date(datas, "dd/MM/yyyy HH:mm")
    from datas_4
""")

display(df_datas_sql3)

df_datas_sql.printSchema()
df_datas_sql2.printSchema()
df_datas_sql3.printSchema()
```
