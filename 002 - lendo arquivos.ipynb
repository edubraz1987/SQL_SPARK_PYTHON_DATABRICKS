```
df_carros = spark.read.format("csv").load("/aprendizado/mode_carro") #carregar o CSV no DataFrame
df_carros = spark.read.format("csv").option("header",True).load("/aprendizado/mode_carro") #carregar com a opção de cabeçalho
df_carros = spark.read.format("csv").option("header",True).option("delimiter",",").load("/aprendizado/mode_carro") #carregar com a opção de informar o separador
df_carros = spark.read.format("csv").option("header",True).load("/aprendizado/mode_carro", sep = "," ) #outra forma de informar o separador
df_carros = spark.read.format("csv").option("header",True).option("enconding","utf-8").load("/aprendizado/mode_carro") #trocar a codificação
display(df_carros)
```

```
#Outros formatos para salvamento do DF
df_carros.write.format("parquet").mode("overwrite").save("/aprendizado/modelo_carro_parquet")
df_carros.write.format("avro").mode("overwrite").save("/aprendizado/modelo_carro_avro")
df_carros.write.format("json").mode("overwrite").save("/aprendizado/modelo_carro_json")
```
