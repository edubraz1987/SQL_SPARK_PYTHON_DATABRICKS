```
from pyspark.sql.functions import *

df_carros = spark.read.format("csv").option("header", True).load("/aprendizado/mode_carro").distinct()

df_carros = df_carros.withColumn("id_carro",col("id_carro").cast("int"))
df_carros = df_carros.where(col("id_carro") <= 4).orderBy("id_carro")

display(df_carros)

df_carros_2 = df_carros.where(
    (col("id_carro")=='1')|
    (col("id_carro")=='2')|
    (col("id_carro")=='3')
)

df_carros_3 = df_carros.where(
    (col("id_carro")=='4')
).withColumn(
    "preco",
    lit(None)
)

df_carros = df_carros_2.union(df_carros_3)
display(df_carros)
```
